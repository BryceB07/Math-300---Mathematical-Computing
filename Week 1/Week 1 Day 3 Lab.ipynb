## Problem 2

Write a block of code that prints out grades. Specifically, given a number `grade`, do the following:

- 'A' if `grade ≥ 90`
- 'B' if `grade ≥ 80`
- 'C' if `grade ≥ 70`
- 'D' if `grade ≥ 60`
- 'F' otherwise

````python
grade = 91
LetterGrade = ""
if grade >= 90:
    LetterGrade = "A"
elif grade >= 80:
    LetterGrade = "B"
elif grade >= 70:
    LetterGrade = "C"
elif grade >= 60:
    LetterGrade = "D"
else:
    LetterGrade = "F"
print(LetterGrade)
````

## Problem 3

A "leap year" is a calendar year where an extra day is added. The rules are as follows:
- If divisible by 400, then leap year
- If divisible by 100, then not a leap year
- If divisible by 4, then leap year
- Otherwise, not a leap year

Write a block of code that takes in the variable `year` and determines if it is a leap year or not.

````python
year = 1900
LeapYear = False
if ((year % 400) == 0):
    LeapYear = True
elif ((year % 100) == 0):
    LeapYear = False
elif ((year % 4) == 0):
    LeapYear = True
else:
    LeapYear = False
print(LeapYear)
````

## Problem 6
Write a for-loop that counts the number of even integers in a list. You can use the following list as an test:

```python
    
    numbers = [1,13,4,15,2,7,8]
```

````python
EvenNumb = []
numbers = [1,13,4,15,2,7,8]
for i in numbers:
    if ((i % 2) == 0):
        EvenNumb.append(i)
print(EvenNumb)
````

## Problem 8
Suppose

```python
a + b = 25
a * b = 144
```

Find all integer solutions `(a, b)`.

---


````python
a = 0
b = 25
numblist = []
for i in range(26):
    if (a * b) == 144:
        numblist.append((a,b))
    a += 1
    b -= 1
print(numblist)
````
